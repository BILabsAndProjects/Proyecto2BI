```python
import pandas as pd

# Cargar los datos
aire = pd.read_csv(r"Datos_aire\Datos POWER BI\SISAIRE-MODIFIED-COMPLETE-IMPUTED-FIXED.csv", sep=';')
sintomas = pd.read_csv(r"Datos_PowerBI/obs_aire_menores5anos_procesado_fixed.csv", sep=';')

# Combinar por año y localidad
merged = pd.merge(aire, sintomas, on=['Año', 'Localidad'], how='inner')

merged.columns
```

```python
merged.head()
```

```python


# Calcular prevalencias
merged['Prevalencia_total'] = merged['Casos sintomas totales'] / merged['Registros']
merged['Sibilancia'] = merged['Casos de sibilancia último año'] / merged['Registros']
merged['Tos'] = merged['Casos de tos último año diferente gripa'] / merged['Registros']
merged['Mocos'] = merged['Casos de mocos o nariz tapada último año'] / merged['Registros']


# Seleccionar columnas relevantes
cols_air = ['CO', 'NO2', 'O3', 'PM10', 'PM25', 'SO2']
cols_sym = ['Sibilancia', 'Tos', 'Mocos', 'Prevalencia_total']

for col in cols_air + cols_sym:
    merged[col] = (
        merged[col]
        .astype(str)           # convierte todo a texto
        .str.replace(',', '.', regex=False)  # cambia coma por punto
        .astype(float)         # convierte a número decimal
    )

# Calcular matriz de correlación
corr = merged[cols_air + cols_sym].corr(method='pearson')

# Extraer solo correlaciones aire vs síntomas
corr_result = corr.loc[cols_air, cols_sym]
print(corr_result)

# (Opcional) exportar a CSV

```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# === 1. Cargar datasets ===
aire = pd.read_csv(r"Datos_aire\Datos POWER BI\SISAIRE-MODIFIED-COMPLETE-IMPUTED-FIXED.csv", sep=';', decimal=",")
sintomas = pd.read_csv(r"Datos_PowerBI/obs_aire_menores5anos_procesado_fixed.csv", sep=';', decimal=",")

# === 2. Asegurar que columnas numéricas sean realmente numéricas ===
aire.columns
```

```python

# === 2. Asegurar que columnas numéricas sean realmente numéricas ===
cols_air = ["CO", "NO2", "O3", "PM10", "PM25", "SO2"]
cols_sym = ["Casos de sibilancia último año", "Casos de tos último año diferente gripa", "Casos de mocos o nariz tapada último año", "Registros"]

for col in cols_air + cols_sym:
    aire[col] = pd.to_numeric(aire[col], errors="coerce") if col in aire.columns else aire[col]
    sintomas[col] = pd.to_numeric(sintomas[col], errors="coerce") if col in sintomas.columns else sintomas[col]

# === 3. Unir datasets ===
merged = pd.merge(aire, sintomas, on=["Año", "Localidad"], how="inner")

# === 4. Calcular prevalencias ===
merged["Prevalencia_sibilancias"] = merged["Casos de sibilancia último año"] / merged["Registros"]
merged["Prevalencia_tos"] = merged["Casos de tos último año diferente gripa"] / merged["Registros"]
merged["Prevalencia_mocos"] = merged["Casos de mocos o nariz tapada último año"] / merged["Registros"]

cols_sym = ["Prevalencia_sibilancias", "Prevalencia_tos", "Prevalencia_mocos"]

# === 5. Calcular correlaciones por localidad y año ===
corr_list = []

for (loc, anio), group in merged.groupby(["Localidad", "Año"]):
    # Escalado dentro del grupo
    scaler = MinMaxScaler()
    scaled = group.copy()
    scaled[cols_air + cols_sym] = scaler.fit_transform(group[cols_air + cols_sym])
    
    # Calcular matriz de correlación
    corr_matrix = scaled[cols_air + cols_sym].corr(method='pearson')
    corr_subset = corr_matrix.loc[cols_air, cols_sym].reset_index()
    corr_subset = corr_subset.melt(id_vars="index", var_name="Sintoma", value_name="Correlacion")
    corr_subset["Localidad"] = loc
    corr_subset["Año"] = anio
    corr_subset.rename(columns={"index": "Contaminante"}, inplace=True)
    corr_list.append(corr_subset)

corr_df = pd.concat(corr_list, ignore_index=True)

# === 6. Verificar rango de correlaciones ===
print("Rango de correlaciones:", corr_df["Correlacion"].min(), "a", corr_df["Correlacion"].max())

# === 7. Guardar CSV final ===
#corr_df.to_csv("correlacion_localidad_anio.csv", index=False, sep=";", decimal=",")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# === 1. Cargar datasets ===
aire = pd.read_csv(r"Datos_aire\Datos POWER BI\SISAIRE-MODIFIED-COMPLETE-IMPUTED-FIXED.csv", sep=';', decimal=",")
sintomas = pd.read_csv(r"Datos_PowerBI/obs_aire_menores5anos_procesado_fixed.csv", sep=';', decimal=",")

# === 2. Definir columnas ===
cols_air = ["CO", "NO2", "O3", "PM10", "PM25", "SO2"]
cols_sym_raw = ["Casos de sibilancia último año", "Casos de tos último año diferente gripa", "Casos de mocos o nariz tapada último año", "Registros"]

# === 3. Convertir solo columnas existentes en cada dataset ===
for df, cols in [(aire, cols_air), (sintomas, cols_sym_raw)]:
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

# === 4. Verificar columnas comunes para hacer el merge ===
if not {"Año", "Localidad"}.issubset(aire.columns) or not {"Año", "Localidad"}.issubset(sintomas.columns):
    raise ValueError("❌ Las columnas 'Año' y 'Localidad' deben existir en ambos archivos.")

# === 5. Unir datasets ===
merged = pd.merge(aire, sintomas, on=["Año", "Localidad"], how="inner")

# === 6. Calcular prevalencias ==
merged["Prevalencia_sibilancias"] = merged["Casos de sibilancia último año"] / merged["Registros"]
merged["Prevalencia_tos"] = merged["Casos de tos último año diferente gripa"] / merged["Registros"]
merged["Prevalencia_mocos"] = merged["Casos de mocos o nariz tapada último año"] / merged["Registros"]



cols_sym = ["Prevalencia_sibilancias", "Prevalencia_tos", "Prevalencia_mocos"]

# === 7. Calcular correlaciones por localidad y año ===
corr_list = []

for (loc, anio), group in merged.groupby(["Localidad", "Año"]):
    # Verifica que haya al menos 2 filas válidas
    if len(group) >= 2:
        # Escalado dentro del grupo
        scaler = MinMaxScaler()
        scaled = group.copy()
        scaled[cols_air + cols_sym] = scaler.fit_transform(group[cols_air + cols_sym])

        # Calcular correlación aire vs síntomas
        corr_matrix = scaled[cols_air + cols_sym].corr(method='pearson')
        corr_subset = corr_matrix.loc[cols_air, cols_sym].reset_index()
        corr_subset = corr_subset.melt(id_vars="index", var_name="Sintoma", value_name="Correlacion")
        corr_subset["Localidad"] = loc
        corr_subset["Año"] = anio
        corr_subset.rename(columns={"index": "Contaminante"}, inplace=True)
        corr_list.append(corr_subset)

# === 8. Concatenar resultados ===
if len(corr_list) == 0:
    print("⚠️ No se pudo calcular ninguna correlación (quizás hay pocos datos por grupo).")
else:
    corr_df = pd.concat(corr_list, ignore_index=True)
    #corr_df.to_csv("correlacion_localidad_anio.csv", index=False, sep=";", decimal=",")
    print("✅ Archivo 'correlacion_localidad_anio.csv' generado correctamente.")
    print(corr_df.head())
```
