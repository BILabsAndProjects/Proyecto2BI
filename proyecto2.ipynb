```python
def process_file(path):
    iboca = pd.read_excel(path, skiprows=4)
    iboca = iboca.iloc[:-3]
    cols = iboca.columns
    new_cols = {'Unnamed: 0': 'Fecha'}
    medida = ['Concentracion', 'NowCast', 'IBOCA']
    #prev_name = cols[1]
    #i = 0
    for col in cols[1:]:
        if 'Unnamed' not in col:
            prev_name = col
            i = 0
        new_name = prev_name + " " + medida[i]
        i += 1
        new_cols[col] = new_name
    iboca = iboca.rename(columns=new_cols)
    iboca = iboca.iloc[2:]
    iboca = iboca.replace('Sin data', 'NaN')
    iboca['Fecha'] = pd.to_datetime(iboca['Fecha'])
    iboca['Año'] = iboca['Fecha'].dt.year

    cols_conc = [c for c in iboca.columns if 'Concentracion' in c]

    iboca_long = iboca.melt(
        id_vars=['Fecha', 'Año'],
        value_vars=cols_conc,
        var_name='Estacion',
        value_name='Concentracion'
    )
    iboca_long['Estacion'] = iboca_long['Estacion'].str.replace(' Concentracion', '', regex=False)
    iboca_long['Localidad'] = iboca_long['Estacion'].map(localidades)
    
    iboca_long = iboca_long[['Estacion', 'Localidad', 'Fecha', 'Año', 'Concentracion']]
    return iboca_long

def save_csv_iboca(pathpm10_1, pathpm25_1, pathpm10_2, pathpm25_2, name):
    pm10_1 = process_file(pathpm10_1)
    pm10_2 = process_file(pathpm10_2)
    pm25_1 = process_file(pathpm25_1)
    pm25_2 = process_file(pathpm25_2)

    pm10 = pd.concat([pm10_1, pm10_2], ignore_index=True)
    pm10 = pm10.rename(columns={'Concentracion': 'Concentracion PM10'})
    
    pm25 = pd.concat([pm25_1, pm25_2], ignore_index=True)
    pm25 = pm25.rename(columns={'Concentracion': 'Concentracion PM2.5'})
    
    
    merged = pd.merge(pm10, pm25, on=['Fecha', 'Año', 'Estacion', 'Localidad'], how='outer')

    merged = merged[['Estacion', 'Localidad', 'Fecha', 'Año', 'Concentracion PM10', 'Concentracion PM2.5']]
    merged.to_csv(name, index=False) 

save_csv_iboca(r'Datos_aire/IBOCA/IBOCA-PM10-2020-1.xlsx', r'Datos_aire/IBOCA/IBOCA-PM25-2020-1.xlsx', r'Datos_aire/IBOCA/IBOCA-PM10-2020-2.xlsx', r'Datos_aire/IBOCA/IBOCA-PM25-2020-2.xlsx','datos_aire_procesados/IBOCA/IBOCA-2020.csv')


```
