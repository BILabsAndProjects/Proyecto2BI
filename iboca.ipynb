```python
df['Estacion'] = df['Estacion'].replace({"Ciudad BolÃ\xadvar": "CIUDAD BOLIVAR", "Ciudad BolГ\xadvar": "CIUDAD BOLIVAR",
                                         "JazmÃ\xadn": "JAZMIN", "JazmГ\xadn": "JAZMIN", "MÃ³vil 7ma": "MOVIL 7MA", 
                                         "MГіvil 7ma": "MOVIL 7MA", "MГіvil FontibГіn": "MOVIL FONTIBON"})
```

```python
df['Estacion'] = (
    df['Estacion']
      .str.upper()
      .str.normalize('NFKD')
      .str.encode('ascii', errors='ignore')
      .str.decode('utf-8')
)
df['Estacion'].unique()
```

```python
localidades = {
    'BOGOTA RURAL - MOCHUELO':'FONTIBON',
    'BOLIVIA': 'ENGATIVA',
    'BOSA':'BOSA',
    'CARVAJAL - SEVILLANA': 'KENNEDY',
    'CDAR':'BARRIOS UNIDOS',
    'CIUDAD BOLIVAR': 'CIUDAD BOLIVAR',
    'COLINA':"SUBA",
    'FONTIBON': 'FONTIBON',
    'GUAYMARAL': 'SUBA',
    'JAZMIN': 'PUENTE ARANDA',
    'KENNEDY': 'KENNEDY',
    'LAS FERIAS': 'ENGATIVA',
    'MINAMBIENTE':'SANTA FE',
    'MOCHUELO - COLEGIO':'CIUDAD BOLIVAR',
    'MOVIL 7MA': 'CHAPINERO',
    'MOVIL FONTIBON':'FONTIBON',
    'P_CAMI - FONTIBON':'FONTIBON',
    'PUENTE ARANDA': 'PUENTE ARANDA',
    'SAN CRISTOBAL': 'SAN CRISTOBAL',
    'SUBA': 'SUBA',
    'TUNAL': 'TUNJUELITO',
    'USAQUEN': 'USAQUEN', 
    'USME': 'USME'
}
df['Localidad'] = df['Estacion'].map(localidades)
df['Localidad'].unique()
```

```python
df[df.duplicated(subset=['Estacion','Fecha'], keep='first')]
df = df.drop_duplicates()
```

```python
df[df.duplicated(subset=['Estacion','Fecha'], keep=False)]
```

```python
# Contar cuántos datos no nulos tiene cada fila
df['non_null_count'] = df.notna().sum(axis=1)

# Para cada Estación + Fecha quedarse con la fila más completa
df = df.sort_values('non_null_count', ascending=False)
df = df.drop_duplicates(subset=['Estacion','Fecha'], keep='first')

# Quitar la columna auxiliar
df = df.drop(columns='non_null_count')
```

```python
df = df.sort_values(['Localidad', 'Estacion', 'Fecha'])

df[['IBOCA PM10','IBOCA PM2.5']] = (
    df.groupby(['Localidad','Estacion'])[['IBOCA PM10','IBOCA PM2.5']]
      .apply(lambda g: g.ffill().bfill())
      .reset_index(level=[0,1], drop=True)
)
```
